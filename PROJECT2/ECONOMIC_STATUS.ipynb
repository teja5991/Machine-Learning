```python
import numpy as np
import random
import pandas as pd
```

```python
#Defining the column which takes categorical values
entity_id=list(range(50000))
city_type = ["A","B","C","D"]
gender = ["male","female"]
marital_status = ["married","unmarried"]
education = ["Ph.d","PG","UG","Not educated"]

#Defining the column which do not take specific values 
#creating list of random integer in the given range 
Dependents=list(range(5))
EntityIncome =list(range(10000,2000000,10000))
#For EntityFamilyIncome column if the number of dependents are 0 then the Entityfamilyincome == Entiyincome this case is taken care of.
#Entityfamilyincome  column is defined later in the code
```

```python
#METHOD OF GENERATING THE DATASET

#If coin flip=1(tail) if coin flip=0(head)
# Rolling the dice if head comes.
#Generating random integer in between 0 and (number of options for given feature -1) 
# meaning the number of options a feature can take
#this process is done for all the columns except for Average Expenditure per Entity per Month column
#As the ques expect us to follow other appproach for generation of that column
```

```python
Dependents_vec=[]
for i in range(50000):
    coin_flip=random.randint(0,1)   #If coin flip=1(tail) if coin flip=0(head)
    if(coin_flip==0):
        dice_roll=random.randint(0,len(Dependents)-1) # Rolling the dice if head comes.
        Dependents_vec.append(Dependents[dice_roll])
    else:
        Dependents_vec.append(2)
```

```python
marital_vec=[]
for i in range(50000):
    coin_flip=random.randint(0,1)     #If coin flip=1(tail) if coin flip=0(head)
    if(coin_flip==0):
        dice_roll=random.randint(0,len(marital_status)-1)# Rolling the dice if head comes.
        marital_vec.append(marital_status[dice_roll])
    else:
        marital_vec.append("married")
```

```python
city_type_vec=[]
for i in range(50000):
    coin_flip=random.randint(0,1)   #If coin flip=1(tail) if coin flip=0(head)
    if(coin_flip==0):
        dice_roll=random.randint(0,len(city_type)-1)# Rolling the dice if head comes.
        city_type_vec.append(city_type[dice_roll])
    else:
        city_type_vec.append("C")
```

```python
gender_vec=[]
for i in range(50000):
    coin_flip=random.randint(0,1)  #If coin flip=1(head) if coin flip=0(tail)
    if(coin_flip==0):
        dice_roll=random.randint(0,len(gender)-1)# Rolling the dice if head comes.
        gender_vec.append(gender[dice_roll])
    else:
        gender_vec.append("female")
```

```python
education_vec=[]
for i in range(50000):
    coin_flip=random.randint(0,1)  #If coin flip=1(head) if coin flip=0(tail)
    if(coin_flip==0):
        dice_roll=random.randint(0,len(education)-1)# Rolling the dice if head comes.
        education_vec.append(education[dice_roll])
    else:
        education_vec.append("UG")
```

```python
entity_income_vec=[]
for i in range(50000):
    coin_flip=random.randint(0,1)   #If coin flip=1(head) if coin flip=0(tail)
    if(coin_flip==0):
        dice_roll=random.randint(0,len(EntityIncome)-1)# Rolling the dice if head comes.
        entity_income_vec.append(EntityIncome[dice_roll])
    else:
        entity_income_vec.append("10000")
```

```python
avg_vec=[]
for i in range(50000):
        if(city_type_vec[i]== "A"):
            avg_vec.append(15000)
        elif (city_type_vec[i]=="D"):
            avg_vec.append(2500)
        elif (city_type_vec[i]=="C"):
            avg_vec.append(5000)
        else:
            avg_vec.append(10000)
```

```python
#For EntityFamilyIncome column if the number of dependents are 0 then the Entityfamilyincome == EntiyIncome this case is taken care of.

entity_family_income_vec=[]

for i in range(50000):
    if(Dependents_vec[i]>0):
        entity_family_income_vec.append(int(entity_income_vec[i])+random.randint(10000,100000))
    else:
        entity_family_income_vec.append(entity_income_vec[i])
```

```python
DataSet = pd.DataFrame()
DataSet["Entity_ID"] = entity_id
DataSet["City"] = city_type_vec
DataSet["Marital Status"] = marital_vec
DataSet["Education"] = education_vec
DataSet["Gender"] =gender_vec
DataSet["Dependencies"]=Dependents_vec
DataSet["Entity's Income"]=entity_income_vec
DataSet["Entity's Family Income"]= entity_family_income_vec
DataSet["Average Expenditure per Entity per Month"]= avg_vec

DataSet
```

```python
# N. Categorize every entity_id economic status as [ Excellent, Good, Moderate, Average, Poor ]

avg_income_vec=[]

for i in range(50000):
    j=random.randint(0,1)
    income=(int(entity_income_vec[i])+int(entity_family_income_vec[i]))
    avg_income_vec.append(income)
                                                       
economic_status_vec=[]   
label_status_vec=[]
for i in range(50000):
    if(avg_income_vec[i] >=900000):
            economic_status_vec.append("Excellent")
            label_status_vec.append(0)
    elif(avg_income_vec[i] >400000 and avg_income_vec[i] < 900000 ):
            economic_status_vec.append("Good")
            label_status_vec.append(1)
    elif(avg_income_vec[i] > 100000 and avg_income_vec[i] <= 400000):
            economic_status_vec.append("Moderate")
            label_status_vec.append(2)
    elif(avg_income_vec[i] >25000 and avg_income_vec[i] <= 100000):
            economic_status_vec.append("Average")    
            label_status_vec.append(3)
    else:
            economic_status_vec.append("Poor") 
            label_status_vec.append(4)
```

```python
DataSet = pd.DataFrame()
DataSet["Entity_ID"] = entity_id
DataSet["City"] = city_type_vec
DataSet["Marital Status"] = marital_vec
DataSet["Education"] = education_vec
DataSet["Gender"] =gender_vec
DataSet["Dependencies"]=Dependents_vec
DataSet["Entity's Income"]=entity_income_vec
DataSet["Entity's Family Income"]= entity_family_income_vec
DataSet["Average Expenditure per Entity per Month"]= avg_vec
DataSet["Total Income of Entity's Family"] = avg_income_vec
DataSet["Economic Status"]= economic_status_vec
DataSet["Label"]=label_status_vec
DataSet
```

```python
# O. Build a prediction model to find the new entity’s Economic status  

temp=DataSet
# Import LabelEncoder
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X=["Entity_ID","City","Marital Status","Education","Gender","Dependencies","Entity\'s Income","Entity\'s Family Income","Average Expenditure per Entity per Month","Total Income of Entity\'s Family"]
Y=["Label"]


le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
temp['Average Expenditure per Entity per Month']=le.fit_transform(temp['Average Expenditure per Entity per Month'])
temp['Gender']=le.fit_transform(temp['Gender'])
temp['Education']=le.fit_transform(temp['Education'])
temp['Marital Status']=le.fit_transform(temp['Marital Status'])
temp['City']=le.fit_transform(temp['City'])

X=temp[X]
Y=temp[Y]

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)


# instantiate the model 
Log_Regression = LogisticRegression(C=10, solver='newton-cg', multi_class='multinomial')

# fit the model with data
Log_Regression.fit(X_train,Y_train)
Y_pred=Log_Regression.predict(X_test)

```

```python


# A. What is the typical shape of the training and testing data ?

print("SIZE OF TRAINING DATASET==>",X_train.size)
print("SIZE OF TESTING DATASET==>",X_test.size)
print("SHAPE OF TRAINING DATASET==>",X_train.shape)
print("SHAPE OF TESTING DATASET==>",X_test.shape)
```

```python

#  P. Perform metric level analysis of the model


from sklearn import metrics

Confusion_Matrix = metrics.confusion_matrix(Y_test, Y_pred)
print(Confusion_Matrix)

FP = Confusion_Matrix .sum(axis=0) - np.diag(Confusion_Matrix )
FN = Confusion_Matrix .sum(axis=1) - np.diag(Confusion_Matrix )
TP = np.diag(Confusion_Matrix)
TN = Confusion_Matrix .sum() - (FP + FN + TP)
from sklearn.metrics import confusion_matrix
import numpy as np


Accuracy=sum(np.diag(Confusion_Matrix ))/ np.sum(Confusion_Matrix )
print("Accuracy in %= ",Accuracy*100)
```

```python
print("Model Evaluation Metrics")
print("Total Observations predicted",len(Y_pred))
print("True Positive=>",TP)
print("True Negative=>",TN)
print("False Positive=>",FP)
print("False Negatives=>",FN)
Accuracy = (TP+TN)/len(Y_pred)
print("Accuracy values for the 5 economic classes = ",Accuracy)
```
